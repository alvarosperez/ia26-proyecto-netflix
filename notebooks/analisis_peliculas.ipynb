```python
import pandas as pd

#with open(...)

df = pd.read_csv("../data/clean/popular_movies.csv")

df.head(1)
```

```python
import matplotlib.pyplot as plt
import ast

generos = df["generos"].str.split(r" \| ").explode().str.strip().value_counts()

plt.figure(figsize=(4, 2))
generos.plot(kind="bar")
```
