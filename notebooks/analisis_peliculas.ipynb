```python
import pandas as pd

# cargar datos de películas
movies_df = pd.read_csv('../data/clean/popular_movies.csv')

movies_df.head(1)
```

```python
import matplotlib.pyplot as plt

# gráfica de géneros de películas
mov_generos = movies_df['genres'].str.split(', ').explode().value_counts()

plt.figure(figsize=(4, 2))
mov_generos.plot(kind='bar')
plt.title('Géneros populares de películas')
plt.show()
```
