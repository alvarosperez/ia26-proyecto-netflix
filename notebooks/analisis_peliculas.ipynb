```python
import pandas as pd

# with open(...)

df = pd.read_csv("../data/clean/popular_movies.csv")
df.head(1)
```

```python
import matplotlib.pyplot as plt

generos = df["genre_ids"].value_counts()

plt.figure(figsize=(10, 6))
generos.plot(kind="bar")
```
