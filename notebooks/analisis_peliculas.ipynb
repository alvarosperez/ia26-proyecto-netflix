```python
import pandas as pd

# with open('../data/netflix_titles.csv', 'r') as f:
#     data = f.read()
data = pd.read_csv('../data/clean/popular_movies.csv')

df = data 

df.head(1)
```

```python
import matplotlib.pyplot as plt

generos = df['genres'].str.split(', ').explode().value_counts()

plt.figure(figsize=(4, 2))
generos.plot(kind='bar')
plt.show()



```
