```python
import pandas as pd # Te faltaba importar pandas
import matplotlib.pyplot as plt
import ast

# 1. Tienes que cargar el archivo CSV que generamos
# Si quieres analizar películas, usa popular_movies.csv
# Si quieres series, usa popular_series.csv
df = pd.read_csv("../data/clean/popular_movies.csv")
# 2. Ahora tu código ya funcionará porque 'df' existe
generos = df["genre_ids"].apply(ast.literal_eval).explode().value_counts()

plt.figure(figsize=(4,2)) # He subido un poco el tamaño para que se lea bien
generos.plot(kind="bar", color="skyblue")
```
